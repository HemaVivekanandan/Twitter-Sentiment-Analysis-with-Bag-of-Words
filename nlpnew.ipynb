```
import pandas as pd
import numpy as np
import keras
import nltk
import sklearn
data = pd.read_csv('train.csv')
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
datatest = pd.read_csv('test.csv')
data
```

```
data['tweet_text'] = data['tweet_text'].str.replace("[^a-zA-Z]", " ")
datatest['tweet_text'] = datatest['tweet_text'].str.replace("[^a-zA-Z]", " ")
data
```

```
nltk.download('punkt')
data['tweet_text']= data['tweet_text'].apply(lambda x: nltk.word_tokenize(x.lower()))
data
```

```
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
data['tweet_text'] = data['tweet_text'].apply(lambda x: [i for i in x if i not in stop_words])
data
```

```
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
ps = WordNetLemmatizer()
data['tweet_text'] = data['tweet_text'].apply(lambda x: [ps.lemmatize(i) for i in x])
data
```

```
texts = data['tweet_text'].to_list()
words = []
for i in texts:
  tokens = i
  for j in tokens:
    words.append(j)
words = nltk.FreqDist(words)
```

```
word_features = list(words.keys())[:1500]
word_features
```

```
def find_features(texts):
  words = []
  for i in texts:
    tokens = i
    for j in tokens:
      words.append(j)

  features = {}
  for i in word_features:
    features[i] = (i in words)
  return features
find_features(texts)
```

```
tweets = list(zip(texts,data['sentiment']))
```

```
featuresets = [(find_features(text),label) for (text, label) in tweets]
featuresets[0]
```

```
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.tree import DecisionTreeClassifier
model = SklearnClassifier(DecisionTreeClassifier())
model.train(featuresets)
```

```
nltk.download('punkt')
datatest['tweet_text']= datatest['tweet_text'].apply(lambda x: nltk.word_tokenize(x.lower()))
datatest
```

```
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
datatest['tweet_text'] = datatest['tweet_text'].apply(lambda x: [i for i in x if i not in stop_words])
datatest
```

```
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
ps = WordNetLemmatizer()
datatest['tweet_text'] = datatest['tweet_text'].apply(lambda x: [ps.lemmatize(i) for i in x])
datatest
```

```
testtext = datatest['tweet_text'].to_list()

def find_features(testtext):
  words = []
  for i in testtext:
    tokens = i
    for j in tokens:
      words.append(j)

  features = {}
  for i in word_features:
    features[i] = (i in words)
  return features
find_features(testtext)
```

```
predlist = []
for text in testtext:
  feature = find_features(text)
  ypred = model.classify(feature)
  predlist.append(ypred)
predlist
```

```
frame = pd.DataFrame({'tweet_id': datatest['tweet_id'],'Sentiment': predlist})
frame
```
